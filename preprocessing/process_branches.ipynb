```python
import pandas as pd
import simplejson
```

```python
df = pd.read_excel('Status ingeleverde data ZH.xlsm', sheet_name='Locaties SEH AV 2021', 
                   skiprows=2, skipfooter=3)
df
```

```python
df = df[['TNO dashboard nr.', '24/7 uurs SEH', 'dag/avond SEH', 'SAZ', 'STZ', 'NFU']]
df = df.replace('x', True).fillna(False)
df['24/7 uurs SEH'] = df['24/7 uurs SEH'].replace(1, True)
df['dag/avond SEH'] = df['dag/avond SEH'].replace(1, True)

df
```

```python
root = {}
for _, row in df.iterrows():
    _id = int(row['TNO dashboard nr.'])
    root[_id] = {
        'fulltimeSEH': row['24/7 uurs SEH'],
        'dagofavondSEH': row['dag/avond SEH'],
        'saz': row['SAZ'],
        'stz': row['STZ'],
        'nfu': row['NFU'],
    }
```

```python
def to_json(dct, filepath):
    with open(filepath, 'w') as f:
        simplejson.dump(dct, f, ignore_nan=True)
to_json(root, 'ziekenhuiscategories.json')
```
