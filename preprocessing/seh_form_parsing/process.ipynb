```python
import seh_form_parser
from pathlib import Path
import os 
import simplejson
from tqdm import tqdm
import json
```

```python
def to_json(dct, filepath):
    with open(filepath, 'w') as f:
        simplejson.dump(dct, f, ignore_nan=True)  # ignore nan -> np.nan to json null's 
```

### All forms

```python
directory = './forms'

excel_filepaths = [Path(os.path.join(directory, file))
                for file in os.listdir(directory) if file.endswith(".xlsx") or file.endswith('.XLSX')]

for excel_filepath in (pbar := tqdm(excel_filepaths)):
    pbar.set_description(f'processing {excel_filepath}')
    json_filepath = excel_filepath.with_suffix('.json')
    parser = seh_form_parser.SEHFormParser(excel_filepath)
    parsed_dict = parser.parse()
    
    filename = os.path.basename(os.path.splitext(excel_filepath)[0])
    _id = int(filename.split('_')[0])
    final_dict = {'id': _id, 'form': parsed_dict}
    to_json(final_dict, json_filepath)
```

#### Combine into one json file

```python
json_files = [os.path.join(directory, file) 
                for file in os.listdir(directory) if file.endswith(".json")]

root = {}
for json_file in json_files:
    with open(json_file) as f:
        filename = os.path.basename(os.path.splitext(json_file)[0])
        
        _id = -1
        if filename != 'Empty':
            _id = int(filename.split('_')[0])

        root[_id] = json.load(f)

to_json(root, 'sehforms.json')
```
