```python
df = pd.read_excel("Pivot SEH AMC uit data CBS.xlsx", sheet_name=2, skiprows=7)
df = df[['Regioletter', '#KaartTNO']]
df
```

```python
dct = {key: value 
       for key, value in zip(df['#KaartTNO'], df['Regioletter'])
       if key != 'x'}
dct
```

```python
import json
with open('knmiregios.json', 'w') as filepath:
    json.dump(dct, filepath)
```
